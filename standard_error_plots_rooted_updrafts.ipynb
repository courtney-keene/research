```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss

# w=0.5 size=4*horiz_thresh
#===============================================
data=pd.read_csv('coh_up_plum_plotter.csv')

ind = data.iloc[1:,0]
N = data.iloc[1:,1]
norm_height = data.iloc[1:,2]

mchord = data.iloc[1:,3]
mdchord = data.iloc[1:,4]
stdchord = data.iloc[1:,5]


chord_std_error = mchord/((N**(1/2)))
mdchord_std_error = mdchord/((N**(1/2)))

WM = data.iloc[1:,6]
WMD = data.iloc[1:,7]
stdW = data.iloc[1:,8]

W_std_error = WM/((N**(1/2)))
Wmd_std_error = WMD/((N**(1/2)))

# this is for a normalized bottom height of less than 0.4
# and for a vertical extent greater than 0.2 
#===========================================================
data_nv5=pd.read_csv('rooted_updrafts_bot.4_ve.2.csv')

ind_nv5 = data_nv5.iloc[1:,0]
N_nv5 = data_nv5.iloc[1:,1]
norm_height_nv5 = data_nv5.iloc[1:,2]

mchord_nv5 = data_nv5.iloc[1:,3]
mdchord_nv5 = data_nv5.iloc[1:,4]
stdchord_nv5 = data_nv5.iloc[1:,5]


chord_std_error_nv5 = mchord_nv5/((N_nv5**(1/2)))
mdchord_std_error_nv5 = mdchord_nv5/((N_nv5**(1/2)))

WM_nv5 = data_nv5.iloc[1:,6]
WMD_nv5 = data_nv5.iloc[1:,7]
stdW_nv5 = data_nv5.iloc[1:,8]

W_std_error_nv5 = WM_nv5/((N_nv5**(1/2)))
Wmd_std_error_nv5 = WMD_nv5/((N_nv5**(1/2)))

# this is for a normalized bottom height of less than 0.2
# and for a vertical extent greater than 0.2 
#===========================================================
data_nv8=pd.read_csv('rooted_updrafts_bot.2_ve.2.csv')

ind_nv8 = data_nv8.iloc[:,0]
N_nv8 = data_nv8.iloc[:,1]
norm_height_nv8 = data_nv8.iloc[:,2]

mchord_nv8 = data_nv8.iloc[:,3]
mdchord_nv8 = data_nv8.iloc[:,4]
stdchord_nv8 = data_nv8.iloc[:,5]

chord_std_error_nv8 = mchord_nv8/((N_nv8**(1/2)))
mdchord_std_error_nv8 = mdchord_nv8/((N_nv8**(1/2)))

WM_nv8 = data_nv8.iloc[:,6]
WMD_nv8 = data_nv8.iloc[:,7]
stdW_nv8 = data_nv8.iloc[:,8]

W_std_error_nv8 = WM_nv8/((N_nv8**(1/2)))
Wmd_std_error_nv8 = WMD_nv8/((N_nv8**(1/2)))
```

```python
plt.figure(figsize=(12,10))
plt.title('Mean Chord Length with Standard Error',fontsize=26)
# plt.title('Median Chord Length with Standard Error',fontsize=24)

plt.errorbar(mchord.iloc[1:],norm_height.iloc[1:],xerr=chord_std_error.iloc[1:],capsize=6,lw=5,elinewidth=3,marker='o',ms=7,
             ecolor='navy',mfc='navy',barsabove=True,label='all updrafts')
plt.errorbar(mchord_nv5.iloc[1:],norm_height_nv5.iloc[1:],xerr=chord_std_error_nv5.iloc[1:],capsize=6,lw=5,elinewidth=3,marker='o',ms=7,
             ecolor='darkred',mfc='darkred',color='indianred',barsabove=True,label='bottom<0.4*Zi'+'\n'+'vert_extent>0.2*Zi')
plt.errorbar(mchord_nv8.iloc[1:],norm_height_nv8.iloc[1:],xerr=chord_std_error_nv8.iloc[1:],capsize=6,lw=5,elinewidth=3,marker='o',ms=7,
             ecolor='darkgreen',mfc='darkgreen',color='mediumseagreen',barsabove=True,label='bottom<0.2*Zi'+'\n'+'vert_extent>0.2*Zi')
# plt.errorbar(mdchord,ind,mdchord_std_error,capsize=6,label='Median and Standard Error')
# plt.legend(fontsize=22)

plt.ylabel('Normalized Height of the BL',fontsize=22)
plt.xlabel('Mean Chord Length [m]',fontsize=22)
# plt.xlabel('Median Chord Length [m]',fontsize=14)
plt.yticks(norm_height[::2],norm_height[::2],fontsize=18)
# plt.xticks(np.arange(75,500,50),fontsize=18)
plt.grid()
plt.legend(fontsize=14)
# plt.savefig('mean_chord_lenght_std_error')
# plt.savefig('median_chord_lenght_std_error')
plt.savefig('powerpoint_plots/mean_chordlength_std_error.png')
plt.show()

#==================================================================================
plt.figure(figsize=(12,10))
plt.title('Mean Maximum Updraft Speed with Standard Error', fontsize=26)
# plt.title('Median Maximum Updraft Speed with Standard Error', fontsize=24)
          
plt.errorbar(WM.iloc[1:],norm_height.iloc[1:],xerr=W_std_error.iloc[1:],capsize=6,lw=5,elinewidth=3,marker='o',ms=7,
             ecolor='navy',mfc='navy',barsabove=True,label='all updrafts')
plt.errorbar(WM_nv5.iloc[1:],norm_height_nv5.iloc[1:],xerr=W_std_error_nv5.iloc[1:],capsize=6,lw=5,elinewidth=3,marker='o',ms=7,color='indianred',
             ecolor='darkred',mfc='darkred',barsabove=True,label='bottom<0.4*Zi'+'\n'+'vert_extent>0.2*Zi')
plt.errorbar(WM_nv8.iloc[5:],norm_height_nv8.iloc[5:],xerr=W_std_error_nv8.iloc[5:],capsize=6,lw=5,elinewidth=3,marker='o',ms=7,
             ecolor='darkgreen',mfc='darkgreen',color='mediumseagreen',barsabove=True,label='bottom<0.2*Zi'+'\n'+'vert_extent>0.2*Zi')
# plt.errorbar(WMD,ind,Wmd_std_error,capsize=6,label='Median and Standard Error')
# plt.legend(fontsize=22)

plt.ylabel('Normalized Height of the BL',fontsize=22)
plt.xlabel('Mean Maximum Vertical Windspeed [m/s]',fontsize=22)
# plt.ylabel('Median Maximum Vertical Windspeed [m/s]',fontsize=14)
plt.yticks(norm_height[::2],norm_height[::2],fontsize=18)
# plt.xticks(np.arange(1.2,2.4,.1),fontsize=18)
plt.grid()
plt.legend(fontsize=14)

# plt.savefig('mean_updraft_std_error_new')
# plt.savefig('median_updraft_std_error')
plt.savefig('powerpoint_plots/mean_updraft_std_error.png')
plt.show()
print(np.where(WM==max(WM)))
```
